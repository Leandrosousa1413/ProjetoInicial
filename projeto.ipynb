```python
quantidade_jogos=dataset['Plataforma'].value_counts()
```

```python
quantidade_jogos=dataset['Plataforma'].value_counts().head(5).plot(kind='bar',figsize=(13,6))
plot.title('Os consoles com mais jogos lançados')
plot.xlabel('Console')
plot.ylabel('Quantidade')
```

```python
dataset[['Nome do Jogo','Vendas Global']].head(10)
```

```python
mais_vendidos=dataset[['Nome do Jogo','Vendas Global']].head(10).set_index('Nome do Jogo').sort_values('Vendas Global', ascending=True).plot(kind='barh')
plot.title('Os jogos mais vendidos')
plot.xlabel('Total de vendas- Em dolares')

plot.show()
                       
```
